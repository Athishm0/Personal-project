```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv('/content/GGR.csv')
df=df.dropna()
df
```

```python
x=df.drop(columns='pIC50',axis=1)
x
```

```python
y=df['pIC50']
y
```

```python
x.shape
```

```python
y.shape
```

```python
#low var fea remove

from sklearn.feature_selection import  VarianceThreshold

selection=VarianceThreshold(threshold=(0.8*(1-0.8)))
x=selection.fit_transform(x)

```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

Regresion Model


```python
y_test
```

```python
import numpy as np
np.random.seed(100)
model=RandomForestRegressor(n_estimators=100)
model.fit(x_train,y_train)
r2=model.score(x_test,y_test)
r2
```

```python
Y_pred=model.predict(x_test)
Y_pred
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_color_codes()
sns.set_style("white")

ax=sns.regplot( y=Y_pred,x=y_test,scatter_kws={'alpha':0.4})
ax.set_xlabel('Exper pIC50',fontsize='large',fontweight='bold')
ax.set_ylabel('Pred pIC50',fontsize='large',fontweight='bold')
#ax.set_xlim(0,12)
#ax.set_ylim(0,12)
ax.figure.set_size_inches(5,5)
plt.show
```
