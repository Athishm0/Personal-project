```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import pyreadr
import feather
from chembl_webresource_client.settings import Settings
```

```python

Settings.Instance().TIMEOUT = 100

target=new_client.target
```

```python
with open("/content/nmk.txt",'r') as inp:
  V=(inp.read()).split(",")
```

```python
X=[]
```

```python

with open("/content/X.txt",'r') as inp:
  X=(inp.read()).split(",")
```

```python
Bad=[]
```

```python
with open("/content/Bad.txt",'r') as inp:
  Bad=(inp.read()).split(",")
```

```python
for i in V:
  if i in Bad or i in X:
    V.remove(i)
```

```python
V
```

```python

for i in V:
   i.rstrip()
   target_q=target.search(i)#.filter(organism='Homo sapiens')

   try:
    targets=(pd.DataFrame.from_dict(target_q))
   except:
    Bad.append(i)
    V.remove(i)
    continue
   try:
      selected_tar=(targets.target_chembl_id[0])
   except:
    Bad.append(i)
    V.remove(i)
    continue
   #if 'standard_type' or 'target_chembl_id'  not in targets:
    # continue

   #for j in range(len(targets.target_chembl_id)):

   act=new_client.activity
   res=act.filter(target_chembl_id=selected_tar).filter(standard_type='IC50')

   df=pd.DataFrame.from_dict(res)
   try:
    df.standard_type.unique()
   except:
    Bad.append(i)
    V.remove(i)
    continue
   X.append(i)
   V.remove(i)

```

```python
AN=iter(V)
while True:
 try:
   i=next(AN)

   i.strip(" ")
   target_q=target.search(i)#.filter(organism='Homo sapiens')

   try:
    targets=(pd.DataFrame.from_dict(target_q))
   except:
    Bad.append(i)
    V.remove(i)
    continue
   try:
      selected_tar=(targets.target_chembl_id[0])
   except:
    Bad.append(i)
    V.remove(i)
    continue
   #if 'standard_type' or 'target_chembl_id'  not in targets:
    # continue

   #for j in range(len(targets.target_chembl_id)):

   act=new_client.activity
   res=act.filter(target_chembl_id=selected_tar).filter(standard_type='IC50')

   df=pd.DataFrame.from_dict(res)
   try:
    df.standard_type.unique()
   except:
    Bad.append(i)
    V.remove(i)
    continue
   X.append(i)
   V.remove(i)
 except StopIteration:
  break



```
