```python
#!pip install biopython

import Bio as bio
from Bio.Seq import Seq
from Bio import Align
from Bio import AlignIO
from Bio import SeqIO


STS=[]
COSMID=[]

aligner=Align.PairwiseAligner(match_score=1.0)
aligner.mode='global'
for sequence in SeqIO.parse("/content/STS.txt","fasta"):
  STS.append(sequence)
for sequence in SeqIO.parse("/content/COSMID.txt","fasta"):
  COSMID.append(sequence)
print(STS)
print(COSMID)



for X in range(len(STS)):

  #for t in range(len(STS[x])
    for i in range(len(COSMID)):
      #for j in range(len(COSMID[i]):
        alignments=aligner.align(COSMID[i],STS[X])
        score=aligner.score(COSMID[i],STS[X])
        if score!=len(STS[X]):
          alignments=aligner.align(COSMID[i],STS[X].reverse_complement)
          score=aligner.score(COSMID[i],STS[X].reverse_complement)
        elif score!=len(STS[X]):
          break

        else:
          for align in alignments:
            print(align)


#AlignIO.write(alignment, "new file", "fasta")

```
